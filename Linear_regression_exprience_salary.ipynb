```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('/home/harsha/Downloads/project_1_dataset.csv',sep=',')
msk = np.random.rand(len(data)) < 0.8
train = data[msk]
test = data[~msk]
train_X = train.iloc[:, 0]
train_Y = train.iloc[:, 1]
test_X = test.iloc[:, 0]
test_Y = test.iloc[:, 1]

plt.scatter(train_X,train_Y)
plt.show()
```

```python
plt.scatter(test_X,test_Y)
plt.show()
```

```python
# Building the model
m = 0
c = 0

L = 0.005  # The learning Rate
epochs = 10000  # The number of iterations to perform gradient descent

n = float(len(train_X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*train_X + c  # The current predicted value of Y
    D_m = (-1/n) * sum(train_X * (train_Y - Y_pred))  # Derivative wrt m
    D_c = (-1/n) * sum(train_Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
```

```python
# Making predictions for train set
Y_pred = m*train_X + c

plt.scatter(train_X, train_Y) 
plt.plot([min(train_X), max(train_X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```

```python
# Making predictions for train set
Y_pred = m*test_X + c

plt.scatter(test_X, test_Y) 
plt.plot([min(test_X), max(test_X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```

```python
predicted =m*15+c
given = 167005.32889087
accuracy =100- abs(((predicted-given)/given)*100)
print("predictes value ",predicted)
print(accuracy)
```
